```python
input_data = senti_data['selected_text']
output_data = senti_data['sentiment']
```

```python
from sklearn.model_selection import train_test_split
input_data_train, input_data_test, output_data_train, output_data_test = train_test_split(input_data,output_data, test_size= 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
model.fit(input_data_train, output_data_train)
```

```python
predicted_senti = model.predict(input_data_test)
```

```python
print(input_data_train.shape)
print(input_data_test.shape)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(output_data_test, predicted_senti)
```

```python
pd.DataFrame(confusion_matrix(output_data_test, predicted_senti), columns = ['Predicted Negative','Predicted_Neutral','Predicted_Positive'], index = ['Actual Negative','Actual Neural','Actual Positive'])
```

```python
from sklearn.metrics import accuracy_score
accuracy_info = accuracy_score(output_data_test, predicted_senti)
print('accuracy :', accuracy_info)
```

```python
def predict_senti(txt, train = input_data_train, model = model):
    pred = model.predict([txt])
    return pred
```
